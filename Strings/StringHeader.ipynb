# String header

Characater pointers (strings) don't really behave like strings until you have access to the string header. The String header contains most of the useful string functions that you'll need to do basic things like comparing two strings, counting the length of a string, adding one string to another string, copying one string into another, and tokenzing.

Let's take a look at how to use these functions with some examples.
Note that we'll #include <string.h> in every example.
